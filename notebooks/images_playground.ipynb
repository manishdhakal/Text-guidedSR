```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
```

```python
# Read coco2017/images/val2017/000000000285.jpg, resize it to 224x224, and display it

image = cv2.imread('images/surfer.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hr_image = cv2.resize(image, (224, 224))

lr_image = cv2.resize(image, (0, 0), fx=1/16, fy=1/16)

interpolated_image = cv2.resize(lr_image, (224, 224), interpolation=cv2.INTER_CUBIC)

# remove axes and display image
plt.axis('off')
diff = hr_image - interpolated_image
diff = diff/255.
noise = np.random.randn(224, 224, 3)
display = diff 
# display = 0.4 * noise
plt.imshow(noise)
```

```python
# Read coco2017/images/val2017/000000000285.jpg, resize it to 224x224, and display it

image = cv2.imread('images/fox.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (224, 224))

# remove axes and display image
plt.axis('off')
plt.imshow(image)
```

```python
# Scale the image down to 8X
image = cv2.resize(image, (0,0), fx=1/16, fy=1/16)

plt.axis('off')
plt.imshow(image)
```

```python
# Interpolate the image back to 224x224
image = cv2.resize(image, (224, 224), interpolation=cv2.INTER_CUBIC)
plt.axis('off')
plt.imshow(image)
```
